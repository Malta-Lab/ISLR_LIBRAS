```python
test_set = pd.read_csv('test_results_model_by_column.csv')
```

```python
test_set
```

```python
## Calculate accuracy, f1, recall and precision for each model from the test_set 
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score

models = test_set.columns[5:]
label = test_set['number_label']
datasets = test_set['dictionary'].unique()

# calculate metrics for each model for each dataset and insert it into a pandas table
metrics = pd.DataFrame(columns=['model', 'dataset', 'accuracy', 'f1', 'recall', 'precision'])

for model in models:
    for dataset in datasets:
        y_pred = test_set[test_set['dictionary'] == dataset][model]
        y_true = test_set[test_set['dictionary'] == dataset]['number_label']
        accuracy = accuracy_score(y_true, y_pred)
        f1 = f1_score(y_true, y_pred, average='macro')
        recall = recall_score(y_true, y_pred, average='macro')
        precision = precision_score(y_true, y_pred, average='macro')
        metrics = metrics.append({'model': model, 'dataset': dataset, 'accuracy': accuracy, 'f1': f1, 'recall': recall, 'precision': precision}, ignore_index=True)


```

```python
metrics[metrics['model'].str.contains('transforms')].sort_values(by='f1', ascending=False)
```

```python
metrics.to_csv('./metrics_for_test_dicts.csv', index=False)
```
